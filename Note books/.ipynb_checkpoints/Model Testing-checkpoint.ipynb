```python
import cv2
import numpy as np
import tensorflow as tf
import seaborn as sns
from matplotlib.animation import FuncAnimation, FFMpegWriter
```

```python
model = tf.keras.models.load_model('../models/best_model_facial_meotion_noramlCNN.h5')

model.summary()
```

```python
faceDetect=cv2.CascadeClassifier('../models/haarcascade_frontalface_default.xml')
```

```python
labels_dict={0:'Angry', 1:'Disgust', 2:'Fear', 3:'Happy', 4:'Neutral', 5:'Sad', 6:'Surprise'}
```

```python
image = cv2.imread('images/small_images.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = faceDetect.detectMultiScale(gray_image, scaleFactor=1.3, minNeighbors=3)

for (x, y, w, h) in faces:
    face_roi = image[y:y + h, x:x + w]
    resized_face = cv2.resize(face_roi, (48, 48))
    normalized_face = resized_face / 255.0
    reshaped_face = np.reshape(normalized_face, (1, 48, 48, 3))
    predicted_emotion = model.predict(reshaped_face)
    emotion_label = np.argmax(predicted_emotion, axis=1)[0]
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 1)
    cv2.rectangle(image, (x, y), (x + w, y + h), (50, 50, 255), 2)
    cv2.rectangle(image, (x, y - 40), (x + w, y), (50, 50, 255), -1)
    cv2.putText(image, labels_dict[emotion_label], (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

cv2.imshow('Emotion Recognition', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()


    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    
    faces = faceDetect.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=3)

    for (x, y, w, h) in faces:

        face_roi = frame[y:y + h, x:x + w]


        resized_face = cv2.resize(face_roi, (48, 48))
        normalized_face = resized_face / 255.0
        reshaped_face = np.reshape(normalized_face, (1, 48, 48, 3))


        predicted_emotion = model.predict(reshaped_face)
        emotion_label = np.argmax(predicted_emotion, axis=1)[0]


        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 1)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (50, 50, 255), 2)
        cv2.rectangle(frame, (x, y - 40), (x + w, y), (50, 50, 255), -1)
        cv2.putText(frame, labels_dict[emotion_label], (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)


    cv2.imshow('Emotion Recognition', frame)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('images/small_images.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = faceDetect.detectMultiScale(gray_image, scaleFactor=1.3, minNeighbors=3)
all_probabilities = []
emotions = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

fig, axs = plt.subplots(len(faces), 2, figsize=(10, 5 * len(faces)))

for i, (x, y, w, h) in enumerate(faces):

    face_roi = image[y:y + h, x:x + w]
    resized_face = cv2.resize(face_roi, (48, 48))
    normalized_face = resized_face / 255.0
    reshaped_face = np.reshape(normalized_face, (1, 48, 48, 3))
    

    predicted_emotion = model.predict(reshaped_face)[0]
    

    all_probabilities.append(predicted_emotion)
    

    axs[i, 0].imshow(cv2.cvtColor(face_roi, cv2.COLOR_BGR2RGB))
    axs[i, 0].axis('off')
    

    axs[i, 1].bar(range(len(emotions)), predicted_emotion, alpha=0.7)
    axs[i, 1].set_title('Emotion Prediction')
    axs[i, 1].set_xticks(range(len(emotions)))
    axs[i, 1].set_xticklabels(emotions)
    axs[i, 1].set_ylabel('Probability')
    plt.savefig('Model Predictions/EmotionalPrediction.png')
plt.tight_layout()


plt.show()
```

```python
from matplotlib.animation import FuncAnimation, PillowWriter

faces = faceDetect.detectMultiScale(gray_image, scaleFactor=1.3, minNeighbors=3)
cap = cv2.VideoCapture('images/Noise.mp4')

fig, axs = plt.subplots(1, 2, figsize=(12, 4))
frame_count=0

def update(frame):
    axs[0].clear()
    axs[1].clear()
    ret, frame = cap.read()
    if not ret:
        return
    
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    frame_count+=1
    faces = faceDetect.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=3)
    
    if len(faces) == 1:
        (x, y, w, h) = faces[0]

        face_roi = frame[y:y + h, x:x + w]
        resized_face = cv2.resize(face_roi, (48, 48))
        normalized_face = resized_face / 255.0
        reshaped_face = np.reshape(normalized_face, (1, 48, 48, 3))
        if frame_count = 10:
            predicted_emotion = model.predict(reshaped_face)[0]
    
    
            cv2.putText(frame, f'Emotion: {np.argmax(predicted_emotion)}', (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)


            axs[0].imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            axs[0].axis('off')
    
            
            x = range(len(predicted_emotion))
            axs[1].bar(x, predicted_emotion, alpha=0.7)
            axs[1].set_title('Emotion Prediction')
            axs[1].set_xticks(x)
            axs[1].set_xticklabels(['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise'])
            axs[1].set_ylabel('Probability')


animation = FuncAnimation(fig, update, frames=100, interval=100)


animation_writer = PillowWriter(fps=10)
animation.save('animated_plot.gif', writer=animation_writer)

cap.release()
```
